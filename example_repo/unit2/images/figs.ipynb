Create a figure for the gradient descent problem:

We have a function whose derivative is:

```
f'(x) = (x-r0)*(x-r2)*(x-r3)
      = x^3 + a2*x^2 + a1*x + a0
```

where

```
a2 = -(r0+r1+r2)
a1 = r0*r1 + r0*r2 + r1*r2
a- = -r0*r1*r2
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Find the coefficients of fgrad = a[0]*x^3 + a[1]*x^2 + a[2]*x + a[3]
roots = np.array([1, 2, 4])
a = np.poly(roots)

# Take the anti-derivative of fgrad to get f
b = np.polyint(a)
b[-1] = 1

# Plot the function f
x  = np.linspace(-0.5, 5, 100)
f = np.polyval(b, x)
plt.plot(x, f)
plt.grid()

# Plot the roots of fgrad
froots = np.polyval(b, roots)
plt.plot(roots, froots, 'ro')

# Draw vertical lines at the roots of fgrad
ymin = -6
ymax = 4
for r, fr in zip(roots, froots):
    plt.plot([r, r], [ymin, fr], color='r', linestyle='--')

plt.ylim([ymin, ymax])
plt.xlabel('x')
plt.ylabel('f(x)')

plt.savefig('func.png')

````
