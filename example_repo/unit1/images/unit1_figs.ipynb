## Graphing Problem

The problem is to graph `f(x) = x*exp(-x)`. The critical points are:

```
f'(x) = exp(-x)*(1-x) = 0 => x = 1.
```

The maximum is `f(1) = 1/e`.  We draw two graphs:  One for the instructor solution and the other for the student solution.

````python
# Instructor solution
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

x = np.linspace(0,5,100)
xmax = 1
ymax = xmax*np.exp(-xmax)
f = x*np.exp(-x)

plt.plot(x, f)
plt.plot([xmax, xmax], [0, ymax], color='r', linestyle='--')
plt.plot([0, xmax], [ymax, ymax], color='r', linestyle='--')
plt.xlim([0, 5])
plt.ylim([0, 0.5])
plt.grid()

ax = plt.gca()
special_x = 1
special_y = np.exp(-1)

xticks = sorted(set(ax.get_xticks()).union([special_x]))
yticks = sorted(set(ax.get_yticks()).union([special_y]))
ax.set_xticks(xticks)
ax.set_yticks(yticks)

y_formatter = ax.yaxis.get_major_formatter()
ax.yaxis.set_major_formatter(
    FuncFormatter(
        lambda value, pos: '1/e' if np.isclose(value, special_y) else y_formatter(value, pos)
    )
)

for tick_value, label in zip(ax.get_xticks(), ax.get_xticklabels()):
    if np.isclose(tick_value, special_x):
        label.set_color('red')

for tick_value, label in zip(ax.get_yticks(), ax.get_yticklabels()):
    if np.isclose(tick_value, special_y):
        label.set_color('red')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.savefig('xexpx_soln.png')
````
